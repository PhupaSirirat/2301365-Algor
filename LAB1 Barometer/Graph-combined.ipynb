```python
# GCD 1
import math
import time
def primeFactors(n): 
    li = []
    while n % 2 == 0:
         li.append(2),
         n  = n / 2 
         #print(n)

    for i in range(3, int(math.sqrt(n))+1, 2):
        while n % i == 0:
            li.append(i),
            n = n / i
            #print("n",n)
    if n > 2:
        li.append(int(n))
    return li

        
def GCD1(m,n):
    ans = 1
    a = primeFactors(m)
    b = primeFactors(n)
    #print(a,b)
    for i in a:
        if i in b:
            #print(i)
            ans = ans*i
            #print(ans)
            b.remove(i) 
    return ans

def count_GCD1(m, n):
    start_time = time.perf_counter()
    GCD1(m, n)
    end_time = time.perf_counter()
    return ('%.10f'%(end_time - start_time))
```

```python
# GCD 2
def SieveOfEratosthenes(num):
    li = []
    prime = [True for i in range(num+1)]
    p = 2
    while (p * p <= num):
        if (prime[p] == True):
            for i in range(p * p, num+1, p):
                prime[i] = False
                
        p += 1
 
    for p in range(2, num+1):
        if prime[p]:
            if num % p == 0:
                li.append(p)
    return factor(num,li)



def factor(num,li):
    new_li=[]
    for i in li:
        while num % i == 0 and num !=0:
            num = num / i
            new_li.append(i)
    return new_li

def GCD2(m,n):
    ans = 1
    a = SieveOfEratosthenes(n)
    b = SieveOfEratosthenes(m)

    for i in a:
        if i in b:                
            ans = ans*i
            b.remove(i) 
    return ans

def count_GCD2(m, n):
    start_time = time.perf_counter()
    GCD2(m, n)
    end_time = time.perf_counter()
    return ('%.10f'%(end_time - start_time))
```
