```python
def find_all_paths(graph):
    all_paths_set = set()  # Set to store unique paths
    all_vertex_pairs = [(i, j) for i in range(len(graph)) for j in range(len(graph)) if i < j]  # Only consider pairs where i < j
    for (i, j) in all_vertex_pairs:
        paths = Path(graph, i, j)
        for p in paths:
            all_paths_set.add(tuple(sorted(p)))  # Add the sorted path tuple to ensure uniqueness
    
    total_paths = len(all_paths_set)
    print(f"Total number of unique paths between all vertex pairs: {total_paths}")

def Path(graph, i, j):
    def dfs(current_node, target_node, visited, path, all_paths):
        visited[current_node] = True
        path.append(current_node)
        
        if current_node == target_node:
            all_paths.append(path.copy())
        else:
            for x in range(len(graph)):
                if graph[current_node][x] == 1 and not visited[x]:
                    dfs(x, target_node, visited, path, all_paths)
        
        path.pop()
        visited[current_node] = False

    visited = [False] * len(graph)
    path = []
    all_paths = []
    
    # Determine if the graph is directed or undirected
    is_directed = is_directed_graph(graph)
    
    # Perform DFS to find all paths
    dfs(i, j, visited, path, all_paths)
    print(f'All paths from {i} to {j}:', all_paths)
    print('The number of paths:', len(all_paths))
    # if all_paths:
    #     for p in all_paths:
    #         visualize_path(graph, p, is_directed)
    # else:
    #     print(f"No path found from {i} to {j}")
    
    return all_paths  # Return the list of paths found for this pair

def is_directed_graph(graph):
    """ Check if the graph is directed based on adjacency matrix symmetry """
    size = len(graph)
    for i in range(size):
        for j in range(i + 1, size):
            if graph[i][j] != graph[j][i]:
                return True
    return False

def visualize_graph(graph, is_directed):
    G = nx.DiGraph() if is_directed else nx.Graph()
    for i in range(len(graph)):
        for j in range(len(graph)):
            if graph[i][j] == 1:
                G.add_edge(i, j)
    
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, arrows=is_directed)
    
    # Show the default graph
    plt.title("Default Directed Graph" if is_directed else "Default Undirected Graph")
    plt.show()

def visualize_path(graph, path, is_directed):
    G = nx.DiGraph() if is_directed else nx.Graph()
    for i in range(len(graph)):
        for j in range(len(graph)):
            if graph[i][j] == 1:
                G.add_edge(i, j)
    
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, arrows=is_directed)
    
    # Highlight the path found
    path_edges = [(path[k], path[k+1]) for k in range(len(path) - 1)]
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='g', width=2, arrows=is_directed)
    nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='g', node_size=700)
    
    # Show the plot with the path highlighted
    plt.title(f"Path: {' -> '.join(map(str, path))}" if is_directed else f"Path: {' - '.join(map(str, path))}")
    plt.show()
```
