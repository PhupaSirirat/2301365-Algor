```python
def analyze_graph(adjacency_matrix, start_node, end_node):
    def is_directed_graph(matrix):
        """Check if the graph is directed based on adjacency matrix symmetry"""
        size = len(matrix)
        for i in range(size):
            for j in range(size):
                if matrix[i][j] != matrix[j][i]:
                    return True
        return False

    def create_graph(matrix, is_directed):
        G = nx.DiGraph() if is_directed else nx.Graph()
        for i in range(len(matrix)):
            for j in range(len(matrix)):
                if matrix[i][j] == 1:
                    G.add_edge(i, j)
        return G

    def has_path(graph, start, end):
        """Check if there is a path from start to end"""
        def dfs(current, target, visited):
            if current == target:
                return True
            visited.add(current)
            for neighbor in graph.neighbors(current):
                if neighbor not in visited:
                    if dfs(neighbor, target, visited):
                        return True
            return False

        visited = set()
        return dfs(start, end, visited)

    def find_all_cycles(graph):
        """Find all cycles in the graph"""
        def find_cycle(node, path, visited, cycles):
            visited.add(node)
            path.append(node)
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    find_cycle(neighbor, path, visited, cycles)
                elif neighbor == path[0] and len(path) > 2:
                    # Valid cycle detected (exclude trivial cycles)
                    cycles.append(path[:] + [neighbor])
            path.pop()
            visited.remove(node)
        
        visited = set()
        cycles = []
        for node in graph.nodes():
            if node not in visited:
                find_cycle(node, [], visited, cycles)
        return cycles

    def visualize_graph(graph, title="Graph", cycles=None):
        """Visualize the graph and highlight any cycles"""
        pos = nx.spring_layout(graph)
        nx.draw(graph, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, arrows=is_directed)
        
        if cycles:
            for cycle in cycles:
                cycle_edges = [(cycle[i], cycle[i + 1]) for i in range(len(cycle) - 1)]
                cycle_edges.append((cycle[-1], cycle[0]))  # Close the cycle
                nx.draw_networkx_edges(graph, pos, edgelist=cycle_edges, edge_color='r', width=2, arrows=is_directed)
                nx.draw_networkx_nodes(graph, pos, nodelist=cycle, node_color='r', node_size=700)
                
            plt.title("Graph with Cycles Highlighted")
        else:
            plt.title(title)
        plt.show()

    # Determine if the graph is directed
    is_directed = is_directed_graph(adjacency_matrix)
    
    # Create the graph
    graph = create_graph(adjacency_matrix, is_directed)
    
    # Visualize the default graph
    visualize_graph(graph, "Default Graph")

    # Check for path
    if has_path(graph, start_node, end_node):
        print(f"There is a path from node {start_node} to node {end_node}.")
    else:
        print(f"There is no path from node {start_node} to node {end_node}.")
    
    # Find all cycles
    cycles = find_all_cycles(graph)
    all_hamil_cyc = []
    if cycles:
        # print("The graph contains cycles:")
        # for cycle in cycles:
        #     print(f"Cycle detected: {cycle}")
        # visualize_graph(graph, "Graph with Cycles Highlighted", cycles)
        # print("The number of cycles in the graph is:", len(cycles))
        # -----------------------------------------------------------
        for cycle in cycles:
            if len(cycle)-1 == len(graph.nodes()):
                print("The graph contains a Hamiltonian cycle.")
                print("Hamiltonian cycle:", cycle)
                all_hamil_cyc.append(cycle)
        print("The number of Hamiltonian cycles in the graph is:", len(all_hamil_cyc))
    else:
        print("The graph does not contain any cycles.")
```
